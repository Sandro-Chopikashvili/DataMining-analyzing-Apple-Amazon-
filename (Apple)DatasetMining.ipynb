```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
df = pd.read_csv('Apple_DB.csv')
plt.figure(figsize=(10, 7))

p = df[df['Year'] == 2022].groupby('Quarter')[['NetSales_Products']].mean()
plt.bar(p.index, p['NetSales_Products'], color='skyblue')

plt.xticks([1,2,3,4])
plt.title('Average Net Sales of Products by Quarter in 2022')
plt.xlabel('Quarters')
plt.ylabel('USD$ in Millions')
```

```python
a = df.groupby('Year')[['NetSales_Total']].mean()
plt.figure(figsize=(10, 7))
plt.bar(a.index, a['NetSales_Total'])
plt.title('Average Net Sales of Products by Year')
plt.xlabel('Years')
plt.ylabel('USD$ in Millions')
```

```python
b =df.groupby('Year')[['Stock_Price']].mean().round()
plt.figure(figsize=(10, 7))
plt.plot(b.index, b['Stock_Price'], c='skyblue', marker='o', linewidth=2, markersize=7)
plt.title('Applec Stock Prices Over The Years')
plt.xlabel('Years')
plt.ylabel('USD$')
```

```python
d = df.groupby('Year')[['Net_income']].mean().round()
plt.figure(figsize=(10, 7))
plt.bar(d.index, d['Net_income'], color='lightgreen')
plt.title('Applec Net Income Over The Years')
plt.xlabel('Years')
plt.ylabel('USD$ in Millions')
```

```python
liabilities_by_year = df.groupby("Year")["Liabilities_Total"].sum()
plt.figure(figsize=(10, 7))
plt.plot(liabilities_by_year.index, liabilities_by_year.values, marker='o', linestyle='-')
plt.title("Total Liabilities by Year")
plt.xlabel("Year")
plt.ylabel("Total Liabilities (in millions)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
